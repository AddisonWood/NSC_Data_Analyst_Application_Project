```python
### These functions are used to parse different tables in an FBRef match report

teamnamematching = {'Atlanta Utd':'Atlanta United', 'Austin FC':'Austin FC','CF Montréal':'CF Montréal','Chicago Fire':'Chicago Fire', 'Colorado Rapids':'Colorado Rapids',
                    'Columbus Crew':'Columbus Crew', 'D.C. United':'D.C. United', 'FC Cincinnati':'FC Cincinnati', 'FC Dallas':'FC Dallas', 'Houston Dynamo':'Houston Dynamo',
                    'Inter Miami':'Inter Miami', 'LA Galaxy':'LA Galaxy', 'Los Angeles FC':'Los Angeles FC', 'Minnesota Utd':'Minnesota United', 'NY Red Bulls':'New York Red Bulls',
                    'NYCFC':'New York City FC', 'Nashville':'Nashville SC', 'New England':'New England Revolution', 'Orlando City':'Orlando City', 'Philadelphia':'Philadelphia Union',
                    'Portland Timbers':'Portland Timbers', 'Real Salt Lake':'Real Salt Lake', 'San Jose':'San Jose Earthquakes', 'Seattle':'Seattle Sounders FC',
                    'Sporting KC':'Sporting KC', 'Toronto FC':'Toronto FC', 'Vancouver':'Vancouver Whitecaps FC',
                    'Montreal Impact':'Montreal Impact'}

def parse_summary_table(summary_table):
    summary_df = pd.DataFrame(columns=['Player', '#', 'Nation', 'Pos', 'Age', 'Min', 'Sub', 'Replace',
                                       'Summary_Performance_Gls', 'Summary_Performance_Ast', 'Summary_Performance_PK', 'Summary_Performance_PKatt', 'Summary_Performance_Sh',
                                       'Summary_Performance_SoT', 'Summary_Performance_CrdY', 'Summary_Performance_CrdR', 'Summary_Performance_Touches', 'Summary_Performance_Press',
                                       'Summary_Performance_Tkl', 'Summary_Performance_Int', 'Summary_Performance_Blocks', 'Summary_Expected_xG', 'Summary_Expected_npxG','Summary_Expected_xA',
                                       'Summary_SCA_SCA', 'Summary_SCA_GCA', 'Summary_Passes_Cmp', 'Summary_Passes_Att', 'Summary_Passes_Cmp%', 'Summary_Passes_Prog',
                                       'Summary_Carries_Carries', 'Summary_Carries_Prog', 'Summary_Dribbles_Succ', 'Summary_Dribbles_Att'], index=range(20))
    rows = summary_table.findAll('tr')
    for r in range(len(rows)):
        row = rows[r]
        player = row.findAll('th')[0].text
        if ('\xa0' in player):
            summary_df['Player'][r] = player.replace('\xa0', '')
            summary_df['Sub'][r] = 1
            summary_df['Replace'][r] = summary_df['Player'][r-1]
            summary_df['Replace'][r-1] = summary_df['Player'][r]
        else:
            summary_df['Player'][r] = player.replace('\xa0', '')
            summary_df['Sub'][r] = 0
        
        summary_df['#'][r] = int(row.findAll('td', {'data-stat':'shirtnumber'})[0].text)
        summary_df['Nation'][r] = row.findAll('td', {'data-stat':'nationality'})[0].text.split(' ')[1]
        summary_df['Pos'][r] = row.findAll('td', {'data-stat':'position'})[0].text
        age = rows[r].findAll('td', {'data-stat':'age'})[0].text.split('-')
        summary_df['Age'][r] = int(age[0]) + int(age[1])/360
        try: summary_df['Min'][r] = int(row.findAll('td', {'data-stat':'minutes'})[0].text)
        except ValueError: summary_df['Min'][r] = np.nan
        if (pd.isna(summary_df['Min'][r])):
            summary_df = summary_df.drop(index=[r])
            continue
        
        summary_df['Summary_Performance_Gls'][r] = int(row.findAll('td', {'data-stat':'goals'})[0].text)
        summary_df['Summary_Performance_Ast'][r] = int(row.findAll('td', {'data-stat':'assists'})[0].text)
        summary_df['Summary_Performance_PK'][r] = int(row.findAll('td', {'data-stat':'pens_made'})[0].text)
        summary_df['Summary_Performance_PKatt'][r] = int(row.findAll('td', {'data-stat':'pens_att'})[0].text)
        summary_df['Summary_Performance_Sh'][r] = int(row.findAll('td', {'data-stat':'shots_total'})[0].text)
        summary_df['Summary_Performance_SoT'][r] = int(row.findAll('td', {'data-stat':'shots_on_target'})[0].text)
        summary_df['Summary_Performance_CrdY'][r] = int(row.findAll('td', {'data-stat':'cards_yellow'})[0].text)
        summary_df['Summary_Performance_CrdR'][r] = int(row.findAll('td', {'data-stat':'cards_red'})[0].text)
        summary_df['Summary_Performance_Press'][r] = int(row.findAll('td', {'data-stat':'pressures'})[0].text)
        summary_df['Summary_Performance_Tkl'][r] = int(row.findAll('td', {'data-stat':'tackles'})[0].text)
        summary_df['Summary_Performance_Int'][r] = int(row.findAll('td', {'data-stat':'interceptions'})[0].text)
        summary_df['Summary_Performance_Blocks'][r] = int(row.findAll('td', {'data-stat':'blocks'})[0].text)
        summary_df['Summary_Expected_xG'][r] = float(row.findAll('td', {'data-stat':'xg'})[0].text)
        summary_df['Summary_Expected_npxG'][r] = float(row.findAll('td', {'data-stat':'npxg'})[0].text)
        summary_df['Summary_Expected_xA'][r] = float(row.findAll('td', {'data-stat':'xa'})[0].text)
        summary_df['Summary_SCA_SCA'][r] = int(row.findAll('td', {'data-stat':'sca'})[0].text)
        summary_df['Summary_SCA_GCA'][r] = int(row.findAll('td', {'data-stat':'gca'})[0].text)
        summary_df['Summary_Passes_Cmp'][r] = int(row.findAll('td', {'data-stat':'passes_completed'})[0].text)
        summary_df['Summary_Passes_Att'][r] = int(row.findAll('td', {'data-stat':'passes'})[0].text)
        try: summary_df['Summary_Passes_Cmp%'][r] = float(row.findAll('td', {'data-stat':'passes_pct'})[0].text)
        except ValueError: summary_df['Summary_Passes_Cmp%'][r] = np.nan
        summary_df['Summary_Passes_Prog'][r] = int(row.findAll('td', {'data-stat':'progressive_passes'})[0].text)
        summary_df['Summary_Carries_Carries'][r] = int(row.findAll('td', {'data-stat':'carries'})[0].text)
        summary_df['Summary_Carries_Prog'][r] = int(row.findAll('td', {'data-stat':'progressive_carries'})[0].text)
        summary_df['Summary_Dribbles_Succ'][r] = int(row.findAll('td', {'data-stat':'dribbles_completed'})[0].text)
        summary_df['Summary_Dribbles_Att'][r] = int(row.findAll('td', {'data-stat':'dribbles'})[0].text)
    summary_df = summary_df.dropna(how='all').reset_index(drop=True)
    return(summary_df)

def parse_passing_table(passing_table):
    passing_df = pd.DataFrame(columns=['Player', '#', 'Nation', 'Pos', 'Age', 'Min', 'Sub', 'Replace',
                                       'Passing_Total_Cmp', 'Passing_Total_Att', 'Passing_Total_Cmp%', 'Passing_Total_TotDist', 'Passing_Total_ProgDist',
                                       'Passing_Short_Cmp', 'Passing_Short_Att', 'Passing_Short_Cmp%', 'Passing_Medium_Cmp', 'Passing_Medium_Att', 'Passing_Medium_Cmp%',
                                       'Passing_Long_Cmp', 'Passing_Long_Att', 'Passing_Long_Cmp%', 'Passing__Ast', 'Passing__xA', 'Passing__KP', 'Passing__1/3',
                                       'Passing__PPA', 'Passing__CrsPA', 'Passing__Prog'], index=range(20))
    rows = passing_table.findAll('tr')
    for r in range(len(rows)):
        row = rows[r]
        player = row.findAll('th')[0].text
        if ('\xa0' in player):
            passing_df['Player'][r] = player.replace('\xa0', '')
            passing_df['Sub'][r] = 1
            passing_df['Replace'][r] = passing_df['Player'][r-1]
            passing_df['Replace'][r-1] = passing_df['Player'][r]
        else:
            passing_df['Player'][r] = player.replace('\xa0', '')
            passing_df['Sub'][r] = 0
        
        passing_df['#'][r] = int(row.findAll('td', {'data-stat':'shirtnumber'})[0].text)
        passing_df['Nation'][r] = row.findAll('td', {'data-stat':'nationality'})[0].text.split(' ')[1]
        passing_df['Pos'][r] = row.findAll('td', {'data-stat':'position'})[0].text
        age = rows[r].findAll('td', {'data-stat':'age'})[0].text.split('-')
        passing_df['Age'][r] = int(age[0]) + int(age[1])/360
        try: passing_df['Min'][r] = int(row.findAll('td', {'data-stat':'minutes'})[0].text)
        except ValueError: passing_df['Min'][r] = np.nan
        if (pd.isna(passing_df['Min'][r])):
            passing_df = passing_df.drop(index=[r])
            continue
        
        passing_df['Passing_Total_Cmp'][r] = int(row.findAll('td', {'data-stat':'passes_completed'})[0].text)
        passing_df['Passing_Total_Att'][r] = int(row.findAll('td', {'data-stat':'passes'})[0].text)
        try: passing_df['Passing_Total_Cmp%'][r] = float(row.findAll('td', {'data-stat':'passes_pct'})[0].text)
        except ValueError: passing_df['Passing_Total_Cmp%'][r] = np.nan
        passing_df['Passing_Total_TotDist'][r] = int(row.findAll('td', {'data-stat':'passes_total_distance'})[0].text)
        passing_df['Passing_Total_ProgDist'][r] = int(row.findAll('td', {'data-stat':'passes_progressive_distance'})[0].text)
        passing_df['Passing_Short_Cmp'][r] = int(row.findAll('td', {'data-stat':'passes_completed_short'})[0].text)
        passing_df['Passing_Short_Att'][r] = int(row.findAll('td', {'data-stat':'passes_short'})[0].text)
        try: passing_df['Passing_Short_Cmp%'][r] = float(row.findAll('td', {'data-stat':'passes_pct_short'})[0].text)
        except ValueError: passing_df['Passing_Short_Cmp%'][r] = np.nan
        passing_df['Passing_Medium_Cmp'][r] = int(row.findAll('td', {'data-stat':'passes_completed_medium'})[0].text)
        passing_df['Passing_Medium_Att'][r] = int(row.findAll('td', {'data-stat':'passes_medium'})[0].text)
        try: passing_df['Passing_Medium_Cmp%'][r] = float(row.findAll('td', {'data-stat':'passes_pct_medium'})[0].text)
        except ValueError: passing_df['Passing_Short_Cmp%'][r] = np.nan
        passing_df['Passing_Long_Cmp'][r] = int(row.findAll('td', {'data-stat':'passes_completed_long'})[0].text)
        passing_df['Passing_Long_Att'][r] = int(row.findAll('td', {'data-stat':'passes_long'})[0].text)
        try: passing_df['Passing_Long_Cmp%'][r] = float(row.findAll('td', {'data-stat':'passes_pct_long'})[0].text)
        except ValueError: passing_df['Passing_Long_Cmp%'][r] = np.nan
        passing_df['Passing__Ast'][r] = int(row.findAll('td', {'data-stat':'assists'})[0].text)
        passing_df['Passing__xA'][r] = float(row.findAll('td', {'data-stat':'xa'})[0].text)
        passing_df['Passing__KP'][r] = int(row.findAll('td', {'data-stat':'assisted_shots'})[0].text)
        passing_df['Passing__1/3'][r] = int(row.findAll('td', {'data-stat':'passes_into_final_third'})[0].text)
        passing_df['Passing__PPA'][r] = int(row.findAll('td', {'data-stat':'passes_into_penalty_area'})[0].text)
        passing_df['Passing__CrsPA'][r] = int(row.findAll('td', {'data-stat':'crosses_into_penalty_area'})[0].text)
        passing_df['Passing__Prog'][r] = int(row.findAll('td', {'data-stat':'progressive_passes'})[0].text)
    passing_df = passing_df.dropna(how='all').reset_index(drop=True)
    return(passing_df)

def parse_passtypes_table(passtypes_table):
    passtypes_df = pd.DataFrame(columns=['Player', '#', 'Nation', 'Pos', 'Age', 'Min', 'Sub', 'Replace',
                                         'Pass Types_Pass Types_Live', 'Pass Types_Pass Types_Dead', 'Pass Types_Pass Types_FK', 'Pass Types_Pass Types_TB',
                                         'Pass Types_Pass Types_Press', 'Pass Types_Pass Types_Sw', 'Pass Types_Pass Types_Crs', 'Pass Types_Pass Types_CK',
                                         'Pass Types_Corner Kicks_In', 'Pass Types_Corner Kicks_Out', 'Pass Types_Corner Kicks_Str',
                                         'Pass Types_Height_Ground', 'Pass Types_Height_Low', 'Pass Types_Height_High',
                                         'Pass Types_Body Parts_Left', 'Pass Types_Body Parts_Right', 'Pass Types_Body Parts_Head', 'Pass Types_Body Parts_TI', 'Pass Types_Body Parts_Other',
                                         'Pass Types_Outcomes_Cmp', 'Pass Types_Outcomes_Off', 'Pass Types_Outcomes_Out', 'Pass Types_Outcomes_Int', 'Pass Types_Outcomes_Blocks'], index=range(20))
    rows = passtypes_table.findAll('tr')
    for r in range(len(rows)):
        row = rows[r]
        player = row.findAll('th')[0].text
        if ('\xa0' in player):
            passtypes_df['Player'][r] = player.replace('\xa0', '')
            passtypes_df['Sub'][r] = 1
            passtypes_df['Replace'][r] = passtypes_df['Player'][r-1]
            passtypes_df['Replace'][r-1] = passtypes_df['Player'][r]
        else:
            passtypes_df['Player'][r] = player.replace('\xa0', '')
            passtypes_df['Sub'][r] = 0
        
        passtypes_df['#'][r] = int(row.findAll('td', {'data-stat':'shirtnumber'})[0].text)
        passtypes_df['Nation'][r] = row.findAll('td', {'data-stat':'nationality'})[0].text.split(' ')[1]
        passtypes_df['Pos'][r] = row.findAll('td', {'data-stat':'position'})[0].text
        age = rows[r].findAll('td', {'data-stat':'age'})[0].text.split('-')
        passtypes_df['Age'][r] = int(age[0]) + int(age[1])/360
        try: passtypes_df['Min'][r] = int(row.findAll('td', {'data-stat':'minutes'})[0].text)
        except ValueError: passtypes_df['Min'][r] = np.nan
        if (pd.isna(passtypes_df['Min'][r])):
            passtypes_df = passtypes_df.drop(index=[r])
            continue
        
        
        passtypes_df['Pass Types_Pass Types_Live'][r] = int(row.findAll('td', {'data-stat':'passes_live'})[0].text)
        passtypes_df['Pass Types_Pass Types_Dead'][r] = int(row.findAll('td', {'data-stat':'passes_dead'})[0].text)
        passtypes_df['Pass Types_Pass Types_FK'][r] = int(row.findAll('td', {'data-stat':'passes_free_kicks'})[0].text)
        passtypes_df['Pass Types_Pass Types_TB'][r] = int(row.findAll('td', {'data-stat':'through_balls'})[0].text)
        passtypes_df['Pass Types_Pass Types_Press'][r] = int(row.findAll('td', {'data-stat':'passes_pressure'})[0].text)
        passtypes_df['Pass Types_Pass Types_Sw'][r] = int(row.findAll('td', {'data-stat':'passes_switches'})[0].text)
        passtypes_df['Pass Types_Pass Types_Crs'][r] = int(row.findAll('td', {'data-stat':'crosses'})[0].text)
        passtypes_df['Pass Types_Pass Types_CK'][r] = int(row.findAll('td', {'data-stat':'corner_kicks'})[0].text)
        passtypes_df['Pass Types_Corner Kicks_In'][r] = int(row.findAll('td', {'data-stat':'corner_kicks_in'})[0].text)
        passtypes_df['Pass Types_Corner Kicks_Out'][r] = int(row.findAll('td', {'data-stat':'corner_kicks_out'})[0].text)
        passtypes_df['Pass Types_Corner Kicks_Str'][r] = int(row.findAll('td', {'data-stat':'corner_kicks_straight'})[0].text)
        passtypes_df['Pass Types_Height_Ground'][r] = int(row.findAll('td', {'data-stat':'passes_ground'})[0].text)
        passtypes_df['Pass Types_Height_Low'][r] = int(row.findAll('td', {'data-stat':'passes_low'})[0].text)
        passtypes_df['Pass Types_Height_High'][r] = int(row.findAll('td', {'data-stat':'passes_high'})[0].text)
        passtypes_df['Pass Types_Body Parts_Left'][r] = int(row.findAll('td', {'data-stat':'passes_left_foot'})[0].text)
        passtypes_df['Pass Types_Body Parts_Right'][r] = int(row.findAll('td', {'data-stat':'passes_right_foot'})[0].text)
        passtypes_df['Pass Types_Body Parts_Head'][r] = int(row.findAll('td', {'data-stat':'passes_head'})[0].text)
        passtypes_df['Pass Types_Body Parts_TI'][r] = int(row.findAll('td', {'data-stat':'throw_ins'})[0].text)
        passtypes_df['Pass Types_Body Parts_Other'][r] = int(row.findAll('td', {'data-stat':'passes_other_body'})[0].text)
        passtypes_df['Pass Types_Outcomes_Cmp'][r] = int(row.findAll('td', {'data-stat':'passes_completed'})[0].text)
        passtypes_df['Pass Types_Outcomes_Off'][r] = int(row.findAll('td', {'data-stat':'passes_offsides'})[0].text)
        passtypes_df['Pass Types_Outcomes_Out'][r] = int(row.findAll('td', {'data-stat':'passes_oob'})[0].text)
        passtypes_df['Pass Types_Outcomes_Int'][r] = int(row.findAll('td', {'data-stat':'passes_intercepted'})[0].text)
        passtypes_df['Pass Types_Outcomes_Blocks'][r] = int(row.findAll('td', {'data-stat':'passes_blocked'})[0].text)
    passtypes_df = passtypes_df.dropna(how='all').reset_index(drop=True)
    return(passtypes_df)

def parse_def_table(def_table):
    def_df = pd.DataFrame(columns=['Player', '#', 'Nation', 'Pos', 'Age', 'Min', 'Sub', 'Replace',
                                   'Defensive Actions_Tackles_Tkl', 'Defensive Actions_Tackles_TklW', 'Defensive Actions_Tackles_Def 3rd', 'Defensive Actions_Tackles_Mid 3rd',
                                   'Defensive Actions_Tackles_Att 3rd', 'Defensive Actions_Vs Dribbles_Tkl', 'Defensive Actions_Vs Dribbles_Att', 'Defensive Actions_Vs Dribbles_Tkl%',
                                   'Defensive Actions_Vs Dribbles_Past', 'Defensive Actions_Pressures_Press', 'Defensive Actions_Pressures_Succ', 'Defensive Actions_Pressures_%',
                                   'Defensive Actions_Pressures_Def 3rd', 'Defensive Actions_Pressures_Mid 3rd', 'Defensive Actions_Pressures_Att 3rd', 'Defensive Actions_Blocks_Blocks',
                                   'Defensive Actions_Blocks_Sh', 'Defensive Actions_Blocks_ShSv', 'Defensive Actions_Blocks_Pass', 'Defensive Actions__Int',
                                   'Defensive Actions__Tkl+Int', 'Defensive Actions__Clr', 'Defensive Actions__Err'], index=range(20))
    rows = def_table.findAll('tr')
    for r in range(len(rows)):
        row = rows[r]
        player = row.findAll('th')[0].text
        if ('\xa0' in player):
            def_df['Player'][r] = player.replace('\xa0', '')
            def_df['Sub'][r] = 1
            def_df['Replace'][r] = def_df['Player'][r-1]
            def_df['Replace'][r-1] = def_df['Player'][r]
        else:
            def_df['Player'][r] = player.replace('\xa0', '')
            def_df['Sub'][r] = 0
        
        def_df['#'][r] = int(row.findAll('td', {'data-stat':'shirtnumber'})[0].text)
        def_df['Nation'][r] = row.findAll('td', {'data-stat':'nationality'})[0].text.split(' ')[1]
        def_df['Pos'][r] = row.findAll('td', {'data-stat':'position'})[0].text
        age = rows[r].findAll('td', {'data-stat':'age'})[0].text.split('-')
        def_df['Age'][r] = int(age[0]) + int(age[1])/360
        try: def_df['Min'][r] = int(row.findAll('td', {'data-stat':'minutes'})[0].text)
        except ValueError: def_df['Min'][r] = np.nan
        if (pd.isna(def_df['Min'][r])):
            def_df = def_df.drop(index=[r])
            continue
        
        def_df['Defensive Actions_Tackles_Tkl'][r] = int(row.findAll('td', {'data-stat':'tackles'})[0].text)
        def_df['Defensive Actions_Tackles_TklW'][r] = int(row.findAll('td', {'data-stat':'tackles_won'})[0].text)
        def_df['Defensive Actions_Tackles_Def 3rd'][r] = int(row.findAll('td', {'data-stat':'tackles_def_3rd'})[0].text)
        def_df['Defensive Actions_Tackles_Mid 3rd'][r] = int(row.findAll('td', {'data-stat':'tackles_mid_3rd'})[0].text)
        def_df['Defensive Actions_Tackles_Att 3rd'][r] = int(row.findAll('td', {'data-stat':'tackles_att_3rd'})[0].text)
        def_df['Defensive Actions_Vs Dribbles_Tkl'][r] = int(row.findAll('td', {'data-stat':'dribble_tackles'})[0].text)
        def_df['Defensive Actions_Vs Dribbles_Att'][r] = int(row.findAll('td', {'data-stat':'dribbles_vs'})[0].text)
        try: def_df['Defensive Actions_Vs Dribbles_Tkl%'][r] = float(row.findAll('td', {'data-stat':'dribble_tackles_pct'})[0].text)
        except ValueError: def_df['Defensive Actions_Vs Dribbles_Tkl%'][r] = np.nan
        def_df['Defensive Actions_Vs Dribbles_Past'][r] = int(row.findAll('td', {'data-stat':'dribbled_past'})[0].text)
        def_df['Defensive Actions_Pressures_Press'][r] = int(row.findAll('td', {'data-stat':'pressures'})[0].text)
        def_df['Defensive Actions_Pressures_Succ'][r] = int(row.findAll('td', {'data-stat':'pressure_regains'})[0].text)
        try: def_df['Defensive Actions_Pressures_%'][r] = float(row.findAll('td', {'data-stat':'pressure_regain_pct'})[0].text)
        except ValueError: def_df['Defensive Actions_Pressures_%'][r] = np.nan
        def_df['Defensive Actions_Pressures_Def 3rd'][r] = int(row.findAll('td', {'data-stat':'pressures_def_3rd'})[0].text)
        def_df['Defensive Actions_Pressures_Mid 3rd'][r] = int(row.findAll('td', {'data-stat':'pressures_mid_3rd'})[0].text)
        def_df['Defensive Actions_Pressures_Att 3rd'][r] = int(row.findAll('td', {'data-stat':'pressures_att_3rd'})[0].text)
        def_df['Defensive Actions_Blocks_Blocks'][r] = int(row.findAll('td', {'data-stat':'blocks'})[0].text)
        def_df['Defensive Actions_Blocks_Sh'][r] = int(row.findAll('td', {'data-stat':'blocked_shots'})[0].text)
        def_df['Defensive Actions_Blocks_ShSv'][r] = int(row.findAll('td', {'data-stat':'blocked_shots_saves'})[0].text)
        def_df['Defensive Actions_Blocks_Pass'][r] = int(row.findAll('td', {'data-stat':'blocked_passes'})[0].text)
        def_df['Defensive Actions__Int'][r] = int(row.findAll('td', {'data-stat':'interceptions'})[0].text)
        def_df['Defensive Actions__Tkl+Int'][r] = int(row.findAll('td', {'data-stat':'tackles_interceptions'})[0].text)
        def_df['Defensive Actions__Clr'][r] = int(row.findAll('td', {'data-stat':'clearances'})[0].text)
        def_df['Defensive Actions__Err'][r] = int(row.findAll('td', {'data-stat':'errors'})[0].text)
    def_df = def_df.dropna(how='all').reset_index(drop=True)
    return(def_df)

def parse_possession_table(possession_table):
    possession_df = pd.DataFrame(columns=['Player', '#', 'Nation', 'Pos', 'Age', 'Min', 'Sub', 'Replace',
                                          'Possession_Touches_Touches', 'Possession_Touches_Def Pen', 'Possession_Touches_Def 3rd', 'Possession_Touches_Med 3rd',
                                          'Possession_Touches_Att 3rd', 'Possession_Touches_Att Pen', 'Possession_Touches_Live', 'Possession_Dribbles_Succ',
                                          'Possession_Dribbles_Att', 'Possession_Dribbles_Succ%', 'Possession_Dribbles_#Pl', 'Possession_Dribbles_Megs',
                                          'Possession_Carries_Carries', 'Possession_Carries_TotDist', 'Possession_Carries_PrgDist', 'Possession_Carries_Prog',
                                          'Possession_Carries_1/3', 'Possession_Carries_CPA', 'Possession_Carries_Mis', 'Possession_Carries_Dis',
                                          'Possession_Receiving_Targ', 'Possession_Receiving_Rec', 'Possession_Receiving_Rec%', 'Possession_Receiving_Prog'], index=range(20))
    rows = possession_table.findAll('tr')
    for r in range(len(rows)):
        row = rows[r]
        player = row.findAll('th')[0].text
        if ('\xa0' in player):
            possession_df['Player'][r] = player.replace('\xa0', '')
            possession_df['Sub'][r] = 1
            possession_df['Replace'][r] = possession_df['Player'][r-1]
            possession_df['Replace'][r-1] = possession_df['Player'][r]
        else:
            possession_df['Player'][r] = player.replace('\xa0', '')
            possession_df['Sub'][r] = 0
        
        possession_df['#'][r] = int(row.findAll('td', {'data-stat':'shirtnumber'})[0].text)
        possession_df['Nation'][r] = row.findAll('td', {'data-stat':'nationality'})[0].text.split(' ')[1]
        possession_df['Pos'][r] = row.findAll('td', {'data-stat':'position'})[0].text
        age = rows[r].findAll('td', {'data-stat':'age'})[0].text.split('-')
        possession_df['Age'][r] = int(age[0]) + int(age[1])/360
        try: possession_df['Min'][r] = int(row.findAll('td', {'data-stat':'minutes'})[0].text)
        except ValueError: possession_df['Min'][r] = np.nan
        if (pd.isna(possession_df['Min'][r])):
            possession_df = possession_df.drop(index=[r])
            continue
        
        possession_df['Possession_Touches_Touches'][r] = int(row.findAll('td', {'data-stat':'touches'})[0].text)
        possession_df['Possession_Touches_Def Pen'][r] = int(row.findAll('td', {'data-stat':'touches_def_pen_area'})[0].text)
        possession_df['Possession_Touches_Def 3rd'][r] = int(row.findAll('td', {'data-stat':'touches_def_3rd'})[0].text)
        possession_df['Possession_Touches_Med 3rd'][r] = int(row.findAll('td', {'data-stat':'touches_mid_3rd'})[0].text)
        possession_df['Possession_Touches_Att 3rd'][r] = int(row.findAll('td', {'data-stat':'touches_att_3rd'})[0].text)
        possession_df['Possession_Touches_Att Pen'][r] = int(row.findAll('td', {'data-stat':'touches_att_pen_area'})[0].text)
        possession_df['Possession_Touches_Live'][r] = int(row.findAll('td', {'data-stat':'touches_live_ball'})[0].text)
        possession_df['Possession_Dribbles_Succ'][r] = int(row.findAll('td', {'data-stat':'dribbles_completed'})[0].text)
        possession_df['Possession_Dribbles_Att'][r] = int(row.findAll('td', {'data-stat':'dribbles'})[0].text)
        try: possession_df['Possession_Dribbles_Succ%'][r] = float(row.findAll('td', {'data-stat':'dribbles_completed_pct'})[0].text)
        except ValueError: possession_df['Possession_Dribbles_Succ%'][r] = np.nan
        possession_df['Possession_Dribbles_#Pl'][r] = int(row.findAll('td', {'data-stat':'players_dribbled_past'})[0].text)
        possession_df['Possession_Dribbles_Megs'][r] = int(row.findAll('td', {'data-stat':'nutmegs'})[0].text)
        possession_df['Possession_Carries_Carries'][r] = int(row.findAll('td', {'data-stat':'carries'})[0].text)
        possession_df['Possession_Carries_TotDist'][r] = int(row.findAll('td', {'data-stat':'carry_distance'})[0].text)
        possession_df['Possession_Carries_PrgDist'][r] = int(row.findAll('td', {'data-stat':'carry_progressive_distance'})[0].text)
        possession_df['Possession_Carries_Prog'][r] = int(row.findAll('td', {'data-stat':'progressive_carries'})[0].text)
        possession_df['Possession_Carries_1/3'][r] = int(row.findAll('td', {'data-stat':'carries_into_final_third'})[0].text)
        possession_df['Possession_Carries_CPA'][r] = int(row.findAll('td', {'data-stat':'carries_into_penalty_area'})[0].text)
        possession_df['Possession_Carries_Mis'][r] = int(row.findAll('td', {'data-stat':'miscontrols'})[0].text)
        possession_df['Possession_Carries_Dis'][r] = int(row.findAll('td', {'data-stat':'dispossessed'})[0].text)
        possession_df['Possession_Receiving_Targ'][r] = int(row.findAll('td', {'data-stat':'pass_targets'})[0].text)
        possession_df['Possession_Receiving_Rec'][r] = int(row.findAll('td', {'data-stat':'passes_received'})[0].text)
        try: possession_df['Possession_Receiving_Rec%'][r] = float(row.findAll('td', {'data-stat':'passes_received_pct'})[0].text)
        except ValueError: possession_df['Possession_Receiving_Rec%'][r] = np.nan
        possession_df['Possession_Receiving_Prog'][r] = int(row.findAll('td', {'data-stat':'progressive_passes_received'})[0].text)
    possession_df = possession_df.dropna(how='all').reset_index(drop=True)
    return(possession_df)

def parse_misc_table(misc_table):
    misc_df = pd.DataFrame(columns=['Player', '#', 'Nation', 'Pos', 'Age', 'Min', 'Sub', 'Replace',
                                    'Miscellaneous_Performance_CrdY', 'Miscellaneous_Performance_CrdR', 'Miscellaneous_Performance_2CrdY',
                                    'Miscellaneous_Performance_Fls', 'Miscellaneous_Performance_Fld', 'Miscellaneous_Performance_Off', 'Miscellaneous_Performance_Crs',
                                    'Miscellaneous_Performance_Int', 'Miscellaneous_Performance_TklW', 'Miscellaneous_Performance_PKwon', 'Miscellaneous_Performance_PKcon',
                                    'Miscellaneous_Performance_OG', 'Miscellaneous_Performance_Recov', 'Miscellaneous_Aerial Duels_Won', 'Miscellaneous_Aerial Duels_Lost',
                                    'Miscellaneous_Aerial Duels_Won%'], index=range(20))
    rows = misc_table.findAll('tr')
    for r in range(len(rows)):
        row = rows[r]
        player = row.findAll('th')[0].text
        if ('\xa0' in player):
            misc_df['Player'][r] = player.replace('\xa0', '')
            misc_df['Sub'][r] = 1
            misc_df['Replace'][r] = misc_df['Player'][r-1]
            misc_df['Replace'][r-1] = misc_df['Player'][r]
        else:
            misc_df['Player'][r] = player.replace('\xa0', '')
            misc_df['Sub'][r] = 0
        
        misc_df['#'][r] = int(row.findAll('td', {'data-stat':'shirtnumber'})[0].text)
        misc_df['Nation'][r] = row.findAll('td', {'data-stat':'nationality'})[0].text.split(' ')[1]
        misc_df['Pos'][r] = row.findAll('td', {'data-stat':'position'})[0].text
        age = rows[r].findAll('td', {'data-stat':'age'})[0].text.split('-')
        misc_df['Age'][r] = int(age[0]) + int(age[1])/360
        try: misc_df['Min'][r] = int(row.findAll('td', {'data-stat':'minutes'})[0].text)
        except ValueError: misc_df['Min'][r] = np.nan
        if (pd.isna(misc_df['Min'][r])):
            misc_df = misc_df.drop(index=[r])
            continue
        
        misc_df['Miscellaneous_Performance_CrdY'][r] = int(row.findAll('td', {'data-stat':'cards_yellow'})[0].text)
        misc_df['Miscellaneous_Performance_CrdR'][r] = int(row.findAll('td', {'data-stat':'cards_red'})[0].text)
        misc_df['Miscellaneous_Performance_2CrdY'][r] = int(row.findAll('td', {'data-stat':'cards_yellow_red'})[0].text)
        misc_df['Miscellaneous_Performance_Fls'][r] = int(row.findAll('td', {'data-stat':'fouls'})[0].text)
        misc_df['Miscellaneous_Performance_Fld'][r] = int(row.findAll('td', {'data-stat':'fouled'})[0].text)
        misc_df['Miscellaneous_Performance_Off'][r] = int(row.findAll('td', {'data-stat':'offsides'})[0].text)
        misc_df['Miscellaneous_Performance_Crs'][r] = int(row.findAll('td', {'data-stat':'crosses'})[0].text)
        misc_df['Miscellaneous_Performance_Int'][r] = int(row.findAll('td', {'data-stat':'interceptions'})[0].text)
        misc_df['Miscellaneous_Performance_TklW'][r] = int(row.findAll('td', {'data-stat':'tackles_won'})[0].text)
        try: misc_df['Miscellaneous_Performance_PKwon'][r] = int(row.findAll('td', {'data-stat':'pens_won'})[0].text)
        except ValueError: misc_df['Miscellaneous_Performance_PKwon'][r] = np.nan
        try: misc_df['Miscellaneous_Performance_PKcon'][r] = int(row.findAll('td', {'data-stat':'pens_conceded'})[0].text)
        except ValueError: misc_df['Miscellaneous_Performance_PKcon'][r] = np.nan
        misc_df['Miscellaneous_Performance_OG'][r] = int(row.findAll('td', {'data-stat':'own_goals'})[0].text)
        misc_df['Miscellaneous_Performance_Recov'][r] = int(row.findAll('td', {'data-stat':'ball_recoveries'})[0].text)
        misc_df['Miscellaneous_Aerial Duels_Won'][r] = int(row.findAll('td', {'data-stat':'aerials_won'})[0].text)
        misc_df['Miscellaneous_Aerial Duels_Lost'][r] = int(row.findAll('td', {'data-stat':'aerials_lost'})[0].text)
        try: misc_df['Miscellaneous_Aerial Duels_Won%'][r] = float(row.findAll('td', {'data-stat':'aerials_won_pct'})[0].text)
        except ValueError: misc_df['Miscellaneous_Aerial Duels_Won%'][r] = np.nan
    misc_df = misc_df.dropna(how='all').reset_index(drop=True)
    return(misc_df)
```

```python
# Iterate through all of the games you have urls for
# There is a 5 second sleep command in addition to the processing time - to be nice to FBRef's servers
# This will likely get tripped up at some point - save what you have, reload, and start from where you left off
for g in matchreport_urls.index[1:]:
    if (len(MLS2021_PlayerGames[(MLS2021_PlayerGames['Date'] == matchreport_urls['Date'][g]) & (MLS2021_PlayerGames['Team'] == teamnamematching[matchreport_urls['Team1'][g]])]) > 0):
        continue
    
    print(matchreport_urls['Date'][g], '--', matchreport_urls['Team1'][g], 'vs.', matchreport_urls['Team2'][g])
    url = matchreport_urls['URL'][g]
    res = requests.get(url)
    ### The next two lines get around the issue with comments breaking the parsing
    comm = re.compile("<!--|-->")
    soup = BeautifulSoup(comm.sub("",res.text),'lxml', from_encoding='utf-8') #lxml

    gen_cols = ['Date', 'Team', 'Opp', 'HorA', 'Manager', 'Captain', 'Referee']

    all_tables = soup.findAll('tbody')

    #team 1
    summary_table = all_tables[0]
    passing_table = all_tables[1]
    passtypes_table = all_tables[2]
    def_table = all_tables[3]
    possession_table = all_tables[4]
    misc_table = all_tables[5]

    summary_df = parse_summary_table(summary_table)
    passing_df = parse_passing_table(passing_table)
    passtypes_df = parse_passtypes_table(passtypes_table)
    def_df = parse_def_table(def_table)
    possession_df = parse_possession_table(possession_table)
    misc_df = parse_misc_table(misc_table)
    df2 = summary_df.merge(right=passing_df, how='inner')
    df3 = df2.merge(right=passtypes_df, how='inner')
    df4 = df3.merge(right=def_df, how='inner')
    df5 = df4.merge(right=possession_df, how='inner')
    team1_df = df5.merge(right=misc_df, how='inner')
    stat_cols = list(team1_df.columns)
    team1_df['Date'] = int(soup.findAll('span', {'class':'venuetime'})[0]['data-venue-date'].replace('-',''))
    team1_df['Team'] = soup.findAll('a', {'itemprop':'name'})[0].text
    team1_df['Opp'] = soup.findAll('a', {'itemprop':'name'})[1].text
    team1_df['HorA'] = 'H'
    team1_df['Manager'] = soup.findAll('div', {'class':'datapoint'})[0].text.split('Manager: ')[1].replace('\xa0', ' ')
    team1_df['Captain'] = soup.findAll('div', {'class':'datapoint'})[1].text.split('Captain: ')[1].replace('\xa0', ' ')
    team1_df['Referee'] = soup.findAll('small')[3].text.split(' (Referee)')[0].replace('\xa0', ' ')
    team1_df = team1_df[gen_cols + stat_cols]

    #team 2
    summary_table = all_tables[7]
    passing_table = all_tables[8]
    passtypes_table = all_tables[9]
    def_table = all_tables[10]
    possession_table = all_tables[11]
    misc_table = all_tables[12]

    summary_df = parse_summary_table(summary_table)
    passing_df = parse_passing_table(passing_table)
    passtypes_df = parse_passtypes_table(passtypes_table)
    def_df = parse_def_table(def_table)
    possession_df = parse_possession_table(possession_table)
    misc_df = parse_misc_table(misc_table)
    df2 = summary_df.merge(right=passing_df, how='inner')
    df3 = df2.merge(right=passtypes_df, how='inner')
    df4 = df3.merge(right=def_df, how='inner')
    df5 = df4.merge(right=possession_df, how='inner')
    team2_df = df5.merge(right=misc_df, how='inner')
    stat_cols = list(team2_df.columns)
    team2_df['Date'] = int(soup.findAll('span', {'class':'venuetime'})[0]['data-venue-date'].replace('-',''))
    team2_df['Team'] = soup.findAll('a', {'itemprop':'name'})[1].text
    team2_df['Opp'] = soup.findAll('a', {'itemprop':'name'})[0].text
    team2_df['HorA'] = 'A'
    team2_df['Manager'] = soup.findAll('div', {'class':'datapoint'})[2].text.split('Manager: ')[1].replace('\xa0', ' ')
    team2_df['Captain'] = soup.findAll('div', {'class':'datapoint'})[3].text.split('Captain: ')[1].replace('\xa0', ' ')
    team2_df['Referee'] = soup.findAll('small')[3].text.split(' (Referee)')[0].replace('\xa0', ' ')
    team2_df = team2_df[gen_cols + stat_cols]

    game_df = pd.concat([team1_df, team2_df]).reset_index(drop=True)
    
    MLS2021_PlayerGames = pd.concat([MLS2021_PlayerGames, game_df]).reset_index(drop=True)
    time.sleep(5)
    
    
```
